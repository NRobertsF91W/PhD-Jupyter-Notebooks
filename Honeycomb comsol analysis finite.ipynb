```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
comsol_df = pd.read_csv("../COMSOL/Data/Honeycomb/finite honeycomb many twists landau.csv", header=4)
comsol_df_old = pd.read_csv("../COMSOL/Data/Honeycomb/finite honeycomb 10radm.csv", header=4)
comsol_df_hexagon = pd.read_csv("../COMSOL/Data/Honeycomb/finite honeycomb 10radm hexagon.csv", header=4)
comsol_df_hexagon_2 = pd.read_csv("../COMSOL/Data/Honeycomb/finite honeycomb 20radm hexagon.csv", header=4)
# comsol_df.iloc[:,3:] = comsol_df.iloc[:,3:]/comsol_df["Norm_const"]

display(comsol_df)
twist_list = []

twist_list_old = []
for twist in comsol_df["% twist (rad/m)"].unique():
    prop_dataframe = comsol_df[comsol_df["% twist (rad/m)"] == twist ]
    twist_list.append(prop_dataframe["Propagation constant (rad/m)"].to_numpy())

for twist in comsol_df_old["% twist (rad/m)"].unique():
    prop_dataframe = comsol_df_old[comsol_df_old["% twist (rad/m)"] == twist ][::2]
    twist_list_old.append(prop_dataframe["Propagation constant (rad/m)"].to_numpy())
```

```python
twist_list[1]
```

```python
fig_1 = plt.figure(figsize=(10,7))
ax_1 = fig_1.add_subplot(111)
avg_list = np.mean(twist_list_old,axis=1)
for n,twist_array in enumerate(twist_list_old):
    twist_rate = comsol_df["% twist (rad/m)"].unique()[n]
    ax_1.scatter(np.arange(len(twist_array)), twist_array-avg_list[n] , label= twist_rate, s=20)

plt.ion()
plt.legend()
plt.show()
```

```python
fig_2 = plt.figure(figsize=(10,7))
ax_2 = fig_2.add_subplot(111)
avg_list = np.mean(twist_list,axis=1)
m=4
twist_rate_n = comsol_df["% twist (rad/m)"].unique()[m]
ax_2.scatter(np.arange(len(twist_list[m])), twist_list[m]-avg_list[m] , label= twist_rate_n, s=20)

plt.ion()
plt.legend()
plt.show()
```
