```python
import numpy as np
import pandas as pd
```

```python
MAX_TWIST = 200*2*np.pi/60
desired_fibre_od = 500e-6
df = pd.DataFrame(columns=[r'Preform OD (mm)', 'Twist Rate (rad/m)', 'Feed Rate (mm per min)'])
df['Preform OD (mm)'] = np.kron(np.arange(6, 11), np.ones(8))
df['Twist Rate (rad/m)'] = np.kron(np.ones(len(np.arange(6, 11))), np.arange(200, 1000, 100))
df['Feed Rate (mm per min)'] = 60*1000*(MAX_TWIST*np.pi*(desired_fibre_od/2)**2)/(np.pi*(df['Preform OD (mm)']*1e-3/2)**2 * df['Twist Rate (rad/m)'])
display(df)
```
