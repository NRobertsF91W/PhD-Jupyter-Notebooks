## Chern Number Calculation 

```python

def u_1_func(band_prop_vals,k1,k2,final_k_step):
    # Here we step along k1 direction and then take the dot product 
    #(braket using conjugate) of the full Ex or Ey vectors
    
    k1_prime = k1+1
    if k1_prime == final_k_step+1: # This just makes sure the data set rolls around when you get to the end
        k1_prime = 0
        
    # 20 steps in k1,k2 - data struct iterates over k2 for each k1
    # This requires us to go up or down in steps of 20 for k1    
    start_column = final_k_step*k1 + k2     
    next_column = final_k_step*k1_prime + k2
    u_temp = np.vdot(band_prop_vals.iloc[:,start_column].to_numpy(),band_prop_vals.iloc[:,next_column].to_numpy())
    u_temp = u_temp/abs(u_temp)
                     
    return u_temp
                     
def u_2_func(band_prop_vals,k1,k2,final_k_step):
    # Here we step along k2 direction and then take the dot product 
    #(braket using conjugate) of the full Ex or Ey vectors
    
    # 20 steps in k1,k2 - data struct iterates over k2 for each k1
    # This requires us to go up or down in steps of 20 for k1

    k2_prime = k2+1
    if k2_prime == final_k_step+1:
        k2_prime = 0
    
    start_column = final_k_step*k1 + k2
    next_column = final_k_step*k1 + k2_prime
    u_temp = np.vdot(band_prop_vals.iloc[:,start_column].to_numpy(), band_prop_vals.iloc[:,next_column].to_numpy())
    u_temp = u_temp/abs(u_temp)
                     
    return u_temp
                     
def f_1_2_func(band_prop_vals,k1,k2,final_k_step):
    U1_k1 = u_1_func(band_prop_vals,k1,k2,final_k_step)
    U2_k2 = u_2_func(band_prop_vals,k1,k2,final_k_step)
    
    k2_prime = k2+1
    k1_prime = k1+1
    
    if k2_prime == final_k_step+1:
        k2_prime = 0
    if k1_prime == final_k_step+1:
        k1_prime = 0
        
    U1_k2 = u_1_func(band_prop_vals,k1,k2_prime,final_k_step)
    U2_k1 = u_2_func(band_prop_vals,k1_prime,k2,final_k_step)
    #print(U1_k1,U2_k2,U1_k2,U2_k1)
    f12 = np.log((U1_k1*U2_k1)/(U2_k2*U1_k2))
    
    return f12
```

```python
def calc_chern_number(k1_range,k2_range,band_data_list,final_k_step):
    chern_numbers = np.zeros(4,dtype=complex)
    
    for i,_band in enumerate(band_data_list):
        #print(_band)
        f12_list = np.zeros((len(k1_range),len(k2_range)),dtype=complex)
        for _k1 in k1_range:
            for _k2 in k2_range:
                f12_list[_k1,_k2] = f_1_2_func(_band,_k1,_k2,final_k_step)
        chern_numbers[i] = np.sum(f12_list)*1/(2*np.pi*1.0j)
        
    print(chern_numbers)
```
