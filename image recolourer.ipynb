```python
def normalise_pic(image):
    """ Takes an input image, 
    then returns a normailsed image with consistent max and min"""
    
    mn = np.min(image)
    mx = np.max(image)
    norm_image = (image - mn) * (255.0 / (mx - mn))
    
    return norm_image
```
